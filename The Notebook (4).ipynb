### Let's answer to some analytical questions 
    
Question 1 - In which year we have the `highest` and `lowest` number of movies?

Now finding out number of movies per year

```python
count_per_year=df.groupby('title_year').size()
```

```python
print(count_per_year.idxmax()) # year with maximum movies
print(count_per_year[count_per_year.idxmax()])
```

```python
print(count_per_year.idxmin()) # year with minimum movies
print(count_per_year[1916])
```

Let's create a line chart to show the result

Question 2 - In which year we have `minimum` and `maximum` average `imdb_score` ?

Let's Use a barchart to visualize the result

```python
ax=avg_scores.plot.bar(figsize=(28,8))
ax.set_ylabel('avgscores')
ax.set_xlabel("year")
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

for column in df.columns:
    if df[column].dtype not in ['int64', 'float64']:
        continue  # Skip non-numeric columns

    # Drop NaN values
    clean_data = df[column].dropna()
    # Plot KDE
    plt.figure(figsize=(5, 3))
    sns.kdeplot(clean_data, shade=True, label=column)
    plt.title(f"KDE Plot for {column}")
    plt.xlabel(column)
    plt.legend()
    plt.show()
```
